```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('customers-1000.csv')
```

```python
df
```

**DATA CLEANING**

```python
df.info()
```

```python
df.iloc[[1,2,4]]
```

```python


# Assuming your DataFrame is called df

# Convert columns to correct data types
df = df.astype({
    "Index": "int64",             # integer
    "Customer Id": "string",      # unique ID, best as string
    "First Name": "string",
    "Last Name": "string",
    "Company": "string",
    "City": "string",
    "Country": "string",
    "Phone 1": "string",
    "Phone 2": "string",
    "Email": "string",
    "Website": "string"
})

# Convert Subscription Date to datetime
df["Subscription Date"] = pd.to_datetime(df["Subscription Date"], errors='coerce')

# Check final datatypes
print(df.dtypes)
```

```python
df["Subscription Year"] = df["Subscription Date"].dt.year
df["Subscription Month"] = df["Subscription Date"].dt.month
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

**Exploratory Data Analysis**

**Which companies have the most customers**

```python
# Top 10 companies with most customers
top_companies = df['Company'].value_counts().head(10)

print("ðŸ¢ Top 10 Companies by Number of Customers:")
print(top_companies)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,4))
sns.barplot(x=top_companies.values, y=top_companies.index, palette='viridis')
plt.title("Top 10 Companies by Number of Customers")
plt.xlabel("Number of Customers")
plt.ylabel("Company")
plt.show()
```

**2. Customers per year**

```python
yearly = df.groupby('Subscription Year')['Customer Id'].count().sort_index()
print(yearly)


```

```python


plt.figure(figsize=(10,8))
sns.barplot(x=yearly.index, y=yearly.values, palette='Blues_d')
plt.title("New Customers per Year")
plt.xlabel("Year")
plt.ylabel("Number of New Customers")
plt.show()
```

**ðŸ“† 3. Customers per month (across all years)**

```python
monthly = df.groupby('Subscription Month')['Customer Id'].count().sort_index()
```

```python
monthly
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=monthly.index, y=monthly.values, palette='coolwarm')
plt.title("Subscriptions by Month (All Years Combined)")
plt.xlabel("Month")
plt.ylabel("Number of New Customers")
plt.show()
```

```python
# Daily or monthly trend
trend = df.groupby(df['Subscription Date'].dt.to_period('M'))['Customer Id'].count()

trend.plot(kind='line', marker='o', figsize=(10,8))
plt.title("Customer Subscription Trend Over Time")
plt.xlabel("Month")
plt.ylabel("New Customers")
plt.grid(True)
plt.show()
```


**ðŸ† Top **20** Countries by Number of Customers ** **bold text**

```python

# Total unique countries
num_countries = df['Country'].nunique()
print(f"ðŸŒ Total unique countries: {num_countries}")

# Top 10 countries by customer count
top_countries = df['Country'].value_counts().head(20)
print("\nðŸ† Top 10 Countries by Number of Customers:")
print(top_countries)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,4))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='Blues_r')
plt.title("Top 10 Countries by Number of Customers")
plt.xlabel("Number of Customers")
plt.ylabel("Country")
plt.show()
```

```python
# Number of unique cities
num_cities = df['City'].nunique()
print(f"ðŸ™ï¸ Total unique cities: {num_cities}")

# Top 10 cities by customer count
top_cities = df['City'].value_counts().head(10)
print("\nðŸ“ Top 10 Cities by Number of Customers:")
print(top_cities)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,4))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='mako')
plt.title("Top 10 Cities by Number of Customers")
plt.xlabel("Number of Customers")
plt.ylabel("City")
plt.show()
```

**Top cities per country**

```python
top_cities_country = (
    df.groupby(['Country', 'City'])
      .size()
      .reset_index(name='Customer Count')
      .sort_values(['Country', 'Customer Count'], ascending=[True, False])
)

print(top_cities_country.head(15))  # top few results
```

```python
city_trend = (
    df.groupby(['Subscription Year', 'City'])
      .size()
      .reset_index(name='Customer Count')
)

top5_cities = df['City'].value_counts().head(5).index

plt.figure(figsize=(8,4))
for city in top5_cities:
    data = city_trend[city_trend['City'] == city]
    plt.plot(data['Subscription Year'], data['Customer Count'], marker='o', label=city)

plt.title("Customer Growth by Year for Top 5 Cities")
plt.xlabel("Year")
plt.ylabel("Customers")
plt.legend()
plt.show()
```
